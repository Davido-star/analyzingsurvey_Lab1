```python
# Creates an empty dictionary for assigning county names with vote counts
counts = {}
with open("KEcounty_votes.txt") as file: # Open and process file
  for line in file:
    line = line.strip() #makes new line at end of each line using strip 
    name, vote = line.split(" - ") #splits line at hyphen and outputs a list with 2 index places, [field 0 ], [field 1]
    if vote not in counts:
      # finds first vote for this county and starts an entry 
      counts[vote] = 1 #assigns votes to countys in dictionary 
    else:
      # add 1 to new instance of the vote count
      counts[vote] = counts[vote] + 1
print(counts)
```

```python
# Create an empty dictionary for assigning county names to vote counts
counts = {}

# Create an empty list with the names of everyone who voted
voted = []

with open("KEcounty_votes.txt") as file:
  for line in file:
    line = line.strip().replace("  "," ") #creates new line at the end of each instance, and replaces extra spaces with uniform spacing
    name, vote = line.split(" - ") #splits the line at the hyphen 
    #clean up the person’s name
    name = name.strip().lower()#converts the data to the same format, replaces areas of two spaces with one  
    if name in voted: # check if this person already voted
      print(name + " has already voted! Fraud!")
      continue
    voted.append(name)   # munge the vote string to clean it up
    vote = vote.strip().lower()
    if not vote in counts: # First vote for this variety
      counts[vote] = 1
    else:       # Increment the vote count
      counts[vote] += 1
print() #this print is empty and adds a return line in the results
print("Results:")

for name in counts:
  print(name + ": " + str(counts[name])) #we want to print the name of the county + how moany votes it was given 
```
